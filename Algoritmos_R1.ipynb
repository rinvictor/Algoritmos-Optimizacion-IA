## Torres de Hanoi con Divide y vencerás

Resuelve el problema de las [Torres de Hanoi](https://es.wikipedia.org/wiki/Torres_de_Han%C3%B3i) para un número arbitrario de fichas utilizando la técnica de divide y vencerás. 

```python
# Torres de Hanoi
def hanoi(n, origen, destino, auxiliar):
    if n == 1: # Si sólo hay un movimiento por realizar se hace directamente de origen a destino
        print(f"Disco 1 de {origen} -> {destino}")
        return
    hanoi(n-1, origen, auxiliar, destino)
    print(f"Disco {n} de {origen} -> {destino}")
    hanoi(n-1, auxiliar, destino, origen)
```

```python
while True:
    try:
        nun_discos = int(input("Número de discos: "))
        if nun_discos > 0:
            break  # Salir del bucle si el número es válido
        else:
            print("Por favor, ingresa un número mayor que cero.")
    except ValueError:
        print("Por favor, ingresa un número válido.")

hanoi(nun_discos, 'Torre 1', 'Torre 2', 'Torre 3')
```

## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos. 

```python
import time 
import matplotlib.pyplot as plt
```

```python
# Sucesión_de_Fibonacci usando un método recursivo
def fibonacci_rec(n):
    if n<=1:
        return n
    return fibonacci_rec(n-1) + fibonacci_rec(n-2)
```

```python
# Sucesión de Fibonacci usando un método iterativo
def fibonacci_iter(n):
    fib = [0, 1]
    
    for i in range(2, n + 1):
        fib.append(fib[i-1] + fib[i-2])

    return fib[n]
```

```python
n_values = list(range(1, 31))  # Prueba para n desde 1 hasta 30
tiempos_rec = []

for n in n_values:
    inicio = time.time()
    resultado = fibonacci_rec(n)
    fin = time.time()
    tiempo_ejecucion = fin - inicio
    tiempos_rec.append(tiempo_ejecucion)

plt.plot(n_values, tiempos_rec, marker='o')
plt.title('Tiempo de ejecución de Fibonacci recursivo.')
plt.xlabel('Número de términos')
plt.ylabel('Tiempo de cómputo (segundos)')
plt.grid(True)
plt.show()
```

```python
n_values = list(range(1, 31))
tiempos_iter = []
for n in n_values:
    start_time = time.time()
    resultado = fibonacci_iter(n)
    elapsed_time = time.time() - start_time
    tiempos_iter.append(elapsed_time)

# Grafica los resultados
plt.plot(n_values, tiempos_iter, marker='o')
plt.title('Tiempo de ejecución de Fibonacci Iterativo')
plt.xlabel('Número de términos')
plt.ylabel('Tiempo de cómputo (segundos)')
plt.grid(True)
plt.show()
```

```python
plt.plot(n_values, tiempos_iter, label='Iterativo', marker='o')
plt.plot(n_values, tiempos_rec, label='Recursivo', marker='o')
plt.title('Comparación de Tiempos de Fibonacci Iterativo y Recursivo')
plt.xlabel('Valor de n')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusiones
Como vemos con la primera aproximación, utilizando un enfoque recursivo, el algoritmo tiene un coste computacional que crece de manera exponencial con una complejidad de O(2^n). Esto se debe a la cantidad de operaciones redundantes que se deben hacer. Si en lugar de utilizar un algoritmo basado en recursividad se usa un algoritmo iterativo el crecimiento es lineal O(n), el tiempo de cómputo crece de manera directamente proporcional a la entrada. Elegir una aproximación iterativa para implementar la sucesión de fibonacci mejora significativamente la eficiencia del algoritmo. 

## Devolución de cambio por técnica voraz

Resuelve el [Problema del Cambio de Moneda](https://es.wikipedia.org/wiki/Problema_de_cambio_de_monedas) utilizando una técnica voraz.

```python
# Problema del cambio de moneda
tipo_monedas = [1, 2, 5, 10, 20, 50, 100, 200] # 100 cnts y 200 cnts para simplificar la conversión de € a céntimos
def cambio_moneda(cantidad_ingresada_cnts):
    resultado = []
    tipo_monedas.sort(reverse=True)
    for moneda in tipo_monedas:
        while cantidad_ingresada_cnts >= moneda:
            resultado.append(moneda)
            cantidad_ingresada_cnts = cantidad_ingresada_cnts - moneda
    return resultado
```

```python
cantidad_ingresada = 945 # En céntimos
resultado = cambio_moneda(cantidad_ingresada)
assert cantidad_ingresada == sum(resultado), "Cantidad ingresada debe coincidir con la suma de las monedas resultado"
print(f"Monedas utilizadas: {resultado}")
```

```python
# Si quisiera trabajar en €
def euros_a_centimos(euros):
    centimos = euros * 100
    return int(centimos)  
```

```python
def representacion_en_euros(resultado):
    representacion_euros = []
    for moneda in resultado:
        if moneda == 200:
            representacion_euros.append("2€")
        elif moneda == 100:
            representacion_euros.append("1€")
        else:
            representacion_euros.append(f"{moneda}ct")

    return representacion_euros
```

```python
cantidad_euros = 25.53
cantidad_centimos = euros_a_centimos(cantidad_euros)
resultado_centimos = cambio_moneda(cantidad_centimos)
resultado_euros = representacion_en_euros(resultado_centimos)
print(f"Moendas utilizadas: {resultado_euros}")
assert cantidad_centimos == sum(resultado_centimos), "Cantidad ingresada debe coincidir con la suma de las monedas resultado"
```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

El problema de las N-reinas consiste en colocar N reinas en un tablero de NxN sin se que se amenacen entre sí. Las reinas se amenazan en el caso de estar en la misma linea, fila o diagonal. Para N=2 y N=3 el problema no tiene solución. 

```python
def es_seguro(tablero, fila, col):
    for i in range(fila):
        if tablero[i] == col or tablero[i] - i == col - fila or tablero[i] + i == col + fila:
            return False
    return True
```

```python
# Comprobación de que está bien:
# Caso 1: Tablero vacío
tablero1 = []
fila1 = 0
col1 = 0
print(es_seguro(tablero1, fila1, col1))  # True

# Caso 2: Una reina en la misma columna
tablero2 = [2, 0]
fila2 = 2
col2 = 0
print(es_seguro(tablero2, fila2, col2))  # False

# Caso 3: Una reina en la misma diagonal
tablero3 = [1, 3]
fila3 = 2
col3 = 2
print(es_seguro(tablero3, fila3, col3))  # False

# Caso 4: Ninguna reina en la misma columna o diagonal
tablero4 = [1, 3]
fila4 = 2
col4 = 0
print(es_seguro(tablero4, fila4, col4))  # True

# Caso 6: Una reina en la misma fila
tablero2 = [2, 0]
fila2 = 2
col2 = 0
print(es_seguro(tablero2, fila2, col2))  # False
```

```python
def colocar_reinas(tablero, fila, n):
    if fila == n: # Se llega a una solución
        print(tablero)
        return

    for col in range(n):
        if es_seguro(tablero, fila, col):
            tablero[fila] = col
            colocar_reinas(tablero, fila + 1, n)
            # backtrack
            tablero[fila] = -1
```

```python
def resolver_n_reinas(n):
    tablero = [-1] * n # Inicializo el tablero a -1 (vacío)
    colocar_reinas(tablero, 0, n)
```

```python
# Imprime todos los posibles resultados
resolver_n_reinas(5)
```

Los resultados se interpretan como:
- En la fila 0, hay una reina colocada en la columna 0
- En la fila 1, hay una reina colocada en la columna 2
- En la fila 2, hay una reina colocada en la columna 4
- En la fila 3, hay una reina colocada en la columna 1
- En la fila 4, hay una reina colocada en la columna 3

Cada array es una posible solución para el N dado.
