Creo una función para poder calcular el costo total:

Para poder generar datos aleatorios y probar el método de fuerza bruta se usa la siguiente función (en el último apartado de test también se usará):

```python
# Generación de datos aleatorios

def generar_datos_aleatorios(num_tomas, num_actores, semilla=None):
    if semilla: # Establecer semilla para reproducibilidad
        random.seed(semilla)
    datos_aleatorios = []
    for _ in range(num_tomas):
        fila = [random.choice([True, False]) for _ in range(num_actores)]
        datos_aleatorios.append(fila)
    return datos_aleatorios

random.seed(42)
numero_de_tomas = random.randint(10, 50)
numero_de_actores = random.randint(2, 20)
```

```python
# Función para calcular el costo total

def calcular_coste_total(tomas, asignacion_escenas):
    dias_unicos = set(asignacion_escenas)

    total = 0
    for dia in dias_unicos:
        indices_escenas = [i for i, x in enumerate(asignacion_escenas) if x == dia]

        coste_dia = tomas[indices_escenas[0]]

        for ind_escena in indices_escenas[1:]:
            coste_dia = [max(x, y) for x, y in zip(coste_dia, tomas[ind_escena])]

        total += sum(coste_dia)

    return total
```

```python
# Función para calcular todas las combinaciones posibles
def combinaciones(escenas, min_tomas, max_tomas):
    num_escenas = len(escenas)
    s = []

    for i in range(min_tomas, max_tomas + 1):
        sol = []
        dia = 1
        for j in range(math.ceil(num_escenas / i)):
            if num_escenas - len(sol) > j:
                sol += [dia] * i
                dia += 1
            elif num_escenas - len(sol) > 0:
                sol += [dia] * (num_escenas - len(sol))
        s.append(sol)

    soluciones = s

    for i in s:
        permutaciones = itertools.permutations(i, num_escenas) # Generar todas las permutaciones posibles
        for p in list(permutaciones):
            if list(p) not in soluciones:
                soluciones.append(list(p))

    return soluciones

def fuerza_bruta(datos, min_al_dia, max_al_dia):
    soluciones = combinaciones(datos, min_al_dia, max_al_dia)

    mejor_sol = []
    menor_coste = float('inf')

    for solucion in soluciones:
        coste_sol = calcular_coste_total(datos, solucion)
        if coste_sol < menor_coste:
            menor_coste = coste_sol
            mejor_sol = solucion

    return menor_coste, mejor_sol
```

```python
def fuerza_bruta_test(datos, min_tomas_al_dia, max_tomas_al_dia):
    menor_coste, mejor_solucion = fuerza_bruta(datos, min_tomas_al_dia, max_tomas_al_dia)
    print(f"\nMenor coste fuerza bruta: {menor_coste}")
    print("Mejor solución:")
    df_mejor_solucion = pd.DataFrame({
        "Toma": [f"Toma {i}" for i in range(1, len(mejor_solucion) + 1)],
        "Día de doblaje": [f"Día {dia}" for dia in mejor_solucion]
    })
    display(df_mejor_solucion)
    print("Jornadas de doblaje:", len(set(mejor_solucion)))

num_escenas = 5
num_actores = 3
min_tomas_al_dia = 1
max_tomas_al_dia = 2

datos_aleatorios_test_fuerza_bruta = generar_datos_aleatorios(num_escenas, num_actores, semilla=42)

fuerza_bruta_test(datos_aleatorios_test_fuerza_bruta, min_tomas_al_dia, max_tomas_al_dia)
```

Calcula la complejidad del algoritmo por fuerza bruta

La función tiene una complejidad de O(n^2), n es el número de elementos que tiene la lista de entrada. Hay un bucle anidado que itera sobre las escenas y en cada iteración se realiza cálculo del coste, con complejidad O(m * n), dependiendo del número de días y de escenas cada día.

```python
def asignar_escenas_a_dias(tomas, max_size, actor=1):
    actor -= 1 # Ajustar el índice del actor a partir de 0
    asignacion_escenas = [0] * len(tomas)

    toma_dia = 0
    dia = 0
    for act in range(len(tomas[0])):
        a = (act + actor) % len(tomas[0])
        for i, toma in enumerate(tomas):
            if toma[a] and asignacion_escenas[i] == 0:
                toma_dia += 1
                asignacion_escenas[i] = dia + 1
                if toma_dia + 1 > max_size:
                    dia += 1
                    toma_dia = 0

    return asignacion_escenas
```

```python
# Explorar soluciones vecinas

def generar_solucion_vecina(c, asignacion_escenas_actual):
    mejor_asignacion_escenas = []
    menor_costo_total = float('inf')

    for indice_escena_actual, escena_actual in enumerate(asignacion_escenas_actual[:-1]):
        for indice_escena_vecina in range(indice_escena_actual + 1, len(asignacion_escenas_actual)):
            if escena_actual != asignacion_escenas_actual[indice_escena_vecina]:
                asignacion_escenas_vecina = asignacion_escenas_actual[:]  # Copiar la asignación de escenas actual
                asignacion_escenas_vecina[indice_escena_actual], asignacion_escenas_vecina[indice_escena_vecina] = asignacion_escenas_vecina[indice_escena_vecina], asignacion_escenas_vecina[indice_escena_actual]  # Intercambiar valores para generar una vecina

                costo_total_vecina = calcular_coste_total(c, asignacion_escenas_vecina)

                if costo_total_vecina <= menor_costo_total:
                    menor_costo_total = costo_total_vecina
                    mejor_asignacion_escenas = asignacion_escenas_vecina

    return menor_costo_total, mejor_asignacion_escenas
```

```python
# Población inicial, posibles soluciones
def generar_poblacion_inicial(asignacion_escenas, min_al_dia, max_al_dia, num_actores):
    poblacion = []
    for r in range(min_al_dia, max_al_dia + 1):
        for a in range(1, num_actores + 1):
            sol = asignar_escenas_a_dias(asignacion_escenas, r, a)
            if sol not in poblacion:
                poblacion.append(sol)
    return poblacion

# Encontrar la mejor solución dentro de la población inicial
def encontrar_mejor_solucion(poblacion, matriz_costos):
    mejor_solucion = []
    menor_costo = float('inf')

    for solucion in poblacion:
        coste_solucion = calcular_coste_total(matriz_costos, solucion)
        if coste_solucion < menor_costo:
            menor_costo = coste_solucion
            mejor_solucion = solucion

        mejor_costo_vecina, mejor_solucion_vecina = generar_solucion_vecina(matriz_costos, solucion)
        if mejor_costo_vecina < menor_costo:
            menor_costo = mejor_costo_vecina
            mejor_solucion = mejor_solucion_vecina

    return menor_costo, mejor_solucion
```

```python
def obtener_mejor_solucion(actores_tomas, min_tomas_al_dia, max_tomas_al_dia):
    num_actores = len(actores_tomas[0])
    poblacion_inicial = generar_poblacion_inicial(actores_tomas, min_tomas_al_dia, max_tomas_al_dia, num_actores)
    print(f"Total soluciones generadas: {len(poblacion_inicial)}")
    menor_costo, mejor_solucion = encontrar_mejor_solucion(poblacion_inicial, actores_tomas)
    return menor_costo, mejor_solucion
```

```python
min_coste, best_sol = obtener_mejor_solucion(df_boolean.values.tolist(), min_tomas_al_dia=min_tomas, max_tomas_al_dia=max_tomas)

data = {"Toma": [f"Toma {i+1}" for i in range(len(best_sol))],
        "Día de doblaje": [f"Día {sol}" for sol in best_sol]}
df_resultado = pd.DataFrame(data)

print("Mejor solución:")
display(df_resultado)

print(f"\nMenor coste: {min_coste}")
print("Jornadas de doblaje:", len(set(best_sol)))
```

(*)Calcula la complejidad del algoritmo

- La función `generar_poblacion_inicial` itera sobre el número de actores y las posibles tomas a grabar al día. En cada iteración, llama a `asignar_escenas_al_dia`. La complejidad de `generar_poblacion_inicial` es entonces O(n * m * (max_tomas_al_dia - min_tomas_al_dia + 1)), donde:
  - `n` es el número de escenas,
  - `m` es el número de actores,
  - `max_tomas_al_dia` es el valor máximo de tomas por día
  - `min_tomas_al_dia` es el valor mínimo de tomas por día.

- La función `encontrar_mejor_solucion` calcula el costo total llamando a `calcular_coste_total`, genera una solución vecina y calcula su costo total. La complejidad depende de estas dos funciones. En el caso de `generar_solucion_vecina`, su complejidad es O(n^2) ya que itera sobre cada escena y por cada escena por las restantes para intercambiarlas. La complejidad de `calcular_coste_total` es O(m * n), ya que depende del número de días y del número de escenas cada día.

- La complejidad total del algoritmo es la complejidad de la función `obtener_mejor_solucion`, que está dominada por `encontrar_mejor_solucion`. Por lo tanto, la complejidad total del algoritmo es aproximadamente de:  
  > O(n^2 * m^2).
